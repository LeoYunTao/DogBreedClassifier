```python
image_generator = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255, rotation_range=40,
                                                                  width_shift_range=0.2, height_shift_range=0.2,
                                                                  shear_range=0.2, zoom_range=0.2, horizontal_flip=True,
                                                                  fill_mode="nearest", validation_split=0.2)
BATCH_SIZE = 128
STEPS_PER_EPOCH = np.ceil((20580*0.2)/BATCH_SIZE)
```

```python
train_generator = image_generator.flow_from_directory(
        r"C:\Users\Leoyu\Desktop\DogBreed\119698_791828_bundle_archive\images\Images",
        target_size=(224, 224),
        batch_size=BATCH_SIZE,
        shuffle=True,
        class_mode='categorical',
        subset='training')

validation_generator = image_generator.flow_from_directory(
        r"C:\Users\Leoyu\Desktop\DogBreed\119698_791828_bundle_archive\images\Images",
        target_size=(224, 224),
        batch_size=BATCH_SIZE,
        shuffle=False,
        class_mode='categorical',
        subset='validation')
```

```python
img = train_generator.filepaths[np.random.randint(low=0, high=train_generator.samples)]
print(img)
plt.imshow(plt.imread(img))
```

```python
model = tf.keras.models.Sequential([
    hub.KerasLayer("https://tfhub.dev/google/tf2-preview/mobilenet_v2/classification/4", input_shape=(224, 224, 3)),
    tf.keras.layers.Dense(1024, activation="relu"),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(120, activation='softmax')
])

model.compile(optimizer=tf.keras.optimizers.Adam(),
              loss=tf.keras.losses.CategoricalCrossentropy(),
              metrics=['acc'])

model.summary()
```

```python
callback = tf.keras.callbacks.EarlyStopping(monitor='val_acc', patience=3)
model.fit(train_generator, validation_data=validation_generator, steps_per_epoch=len(train_generator), 
          validation_steps=len(validation_generator) , epochs=25,  callbacks=[callback])
```

```python
imageno=np.random.randint(low=0, high=validation_generator.samples)

name = validation_generator.filepaths[imageno]
print(name)
plt.imshow(plt.imread(name))

img = tf.image.resize(np.asarray(plt.imread(name)), (224, 224)) / 255.0
probabilities = model.predict(np.expand_dims(img, axis=0))
breed_list = tuple(zip(validation_generator.class_indices.values(), validation_generator.class_indices.keys()))

for i in probabilities[0].argsort()[-5:][::-1]: 
    print(probabilities[0][i], "  :  " , breed_list[i])
```
